```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

#  ,   
node_list = [1, 2, 3, 4, 5]
edge_list = [
    (1, 1), (1, 2), (1, 3), (1, 4), (1, 5),
    (2, 1), (2, 3), (2, 5),
    (3, 1), (3, 2),
    (4, 1),
    (5, 1), (5, 2)
]

#   (networkx  )
G = nx.Graph()
G.add_nodes_from(node_list)
G.add_edges_from(edge_list)

#  
adj_matrix = nx.adjacency_matrix(G).todense()
print("Adjacency Matrix:")
print(adj_matrix)

#  
custom_labels = {
    1: "Seoul", 2: "Daejeon", 3: "Daegu", 4: "Busan", 5: "Gwangju"
}

#  
plt.figure(figsize=(6, 6))
nx.draw(G, with_labels=True, labels=custom_labels, node_color='lightblue', node_size=1500, font_size=16, font_weight='bold')
plt.title("Big Cities in Korea")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

#    
G = nx.karate_club_graph()
```

```python
#   /  
print(f"Nodes: {G.number_of_nodes()}")
print(f"Edges: {G.number_of_edges()}")
```

```python
G.nodes
```

```python
# id 100    
G.add_node(100)
```

```python
G.nodes
```

```python
G.edges
```

```python
# Connect the new node to all existing nodes
#    
for node in G.nodes:
    if node != 100:  # Avoid self-loop
        G.add_edge(100, node, weight=1)
```

```python
G.edges
```

```python
#   ->    
G.remove_node(100)
```

```python
G.edges
```

```python
#  
plt.figure(figsize=(10, 10))
pos = nx.circular_layout(G)
nx.draw(G, pos, with_labels=True, node_size=1000, font_size=12, font_weight='bold')
plt.title("Karate Club Graph Visualization")
plt.show()
```

```python
#   
degree_sequence = [d for n, d in G.degree()]
print(f"Degree sequence: {degree_sequence}")
```

```python
[(n, d) for n, d in G.degree()]
```

```python
#    
average_degree = np.mean(degree_sequence)
print(f"Average degree: {average_degree}")
```

```python
# / 
max_degree = np.max(degree_sequence)
min_degree = np.min(degree_sequence)
print(f"Max degree: {max_degree}, Min degree: {min_degree}")
```

```python
#  
plt.hist(degree_sequence, bins=range(min(degree_sequence), max(degree_sequence) + 2))
plt.title("Degree Histogram")
plt.xlabel("Degree")
plt.ylabel("Frequency")
plt.show()
```

```python
#  ?
density = nx.density(G)
print(f"Density: {density}")
```

```python
#       
path_length = nx.shortest_path_length(G, source=1, target=22)
print(f"Shortest path length from node 1 to node 10: {path_length}")
```

```python
#  
A = nx.adjacency_matrix(G)
print("Adjacency matrix:")
print(A.todense())
```

```python
import networkx as nx
from community import community_louvain
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

#   
G = nx.les_miserables_graph()

# Louvain    partition 
partition = community_louvain.best_partition(G)

#   
colors = [mcolors.CSS4_COLORS[key] for key in mcolors.CSS4_COLORS.keys()][20:40]
node_colors = [colors[community] for community in partition.values()]

# spring layout  
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_color=node_colors, node_size=100, font_size=8, font_weight='bold')
plt.title("Les Misrables Graph Visualization")
plt.show()
```
